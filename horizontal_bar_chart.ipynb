## Horizontal bar chart from a counter (Matplotlib)

```python
import matplotlib.pyplot as plt
from collections import Counter

data = ['spam', 'spam', 'bacon', 'spam', 'sausage', 'eggs', 'eggs', 'ham', 'ham', 'Lobster Thermidor au Crevette']
counter = Counter(data)

# sort items by frequency, use most_common() to show all
#labels, values = map(list, zip(*counter.most_common(5)))
labels, values = zip(*counter.most_common(5))
print(labels, values)
y_pos = range(len(labels))

# size and style
fig = plt.figure(figsize=(20, 15))
plt.style.use('ggplot')
ax = plt.gca()

# labels
ax.set_yticks(y_pos)
ax.set_yticklabels(labels)
ax.invert_yaxis()

# label font size
plt.tick_params(labelsize=20)

# plot
ax.barh(y_pos, values, align='center')
plt.show()
```
